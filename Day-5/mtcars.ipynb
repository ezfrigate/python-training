```python
import pandas as pd
import statsmodels.api as sm

df = pd.read_csv('C:/Local/python-training/Day-5/mtcars.csv')
df
```

```python
#Linear Regression

#X = df[['cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']]  # Independent variables
X = df[['hp','gear']]  # Independent variables
X = sm.add_constant(X)  # Add a constant (intercept) to the model
y = df['mpg']  # Dependent variable

model = sm.OLS(y, X).fit()  # Fit Ordinary Least Squares (OLS) regression model
model.summary()
```

```python
import statsmodels.formula.api as smf

model = smf.ols(formula= 'mpg ~ cyl+disp+hp+drat+wt+qsec+vs+am+gear+carb', data = df)
result = model.fit()
result.summary()
```

```python
import matplotlib.pyplot as plt

plt.rc("figure", figsize=(16, 36))
plt.rc("font", size=14)
fig = sm.graphics.plot_partregress_grid(result)
fig.tight_layout(pad=1.0)
```
