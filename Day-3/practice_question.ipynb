```python
#convert all ones in any case to 1 and twos in any case to 2
import pandas
df = pandas.read_csv('/content/student.csv')
df
```

```python
df.Section.unique()
```

```python
df.Section = df.Section.map({'One' : 1, 'Two' : 2})
df.Section.unique()
```

```python
#I want the percentages of all the students in a separate column called percentage

df['Percentage'] = (df.Maths + df.Science + df.English)/3
df.head()
```

```python
#I want to filter class 1 and class 2 students seperately and store in two different data frames df1 and df2

df1 = df[(df.Section == 1)]
df2 = df[(df.Section == 2)]
df2.head()
```

```python
#I want topper of class 1 in maths

df1[(df1.Maths == df1.Maths.max())]
```

```python
#I want topper of class 2 in English

df2[(df2.English == df2.English.max())]
```

```python
#I want overall topper in class 1
#I want overall topper in class 2

df1[(df1.Percentage == df1.Percentage.max())]
df2[(df2.Percentage == df2.Percentage.max())]
```

```python
#I want names of the students which starts with A along with their classes
#I want names of the students which ends with n along with their classes

df[(df.Name.str.startswith('A'))][['Name', 'Section']]
df[(df.Name.str.endswith('n'))][['Name', 'Section']]
```

```python
#I want names of all the students who failed in class 1 in any of the subjects (fail means less than 40% marks in a subject)

df1[(df1.Maths<40) | (df1.Science<40) | (df1.English<40)][['Name']]
```

```python
#I want names of the students in class 1 who failed in all the subjects

df1[(df1.Maths<40) & (df1.Science<40) & (df1.English<40)][['Name']]
```

```python
#I want name and class of students who failed in maths but passed in English and Science

df[(df.Maths<40) & (df.Science>=40) & (df.English>=40)][['Name','Section']]
```
