```python
'''
Pandas provides several types of plots through its built-in plotting functionality. Here's a list of some common plot types available in Pandas:

1. **Line Plot (`kind='line'`)**: Line plots are used to visualize data points connected by lines. You can create line plots using `df.plot()` with `kind='line'`.

2. **Bar Plot (`kind='bar'` or `kind='barh'`)**: Bar plots are used to represent categorical data with rectangular bars. Vertical and horizontal bar plots can be created using `df.plot(kind='bar')` or `df.plot(kind='barh')`.

3. **Histogram (`kind='hist'`)**: Histograms display the distribution of a continuous variable. You can create histograms using `df.plot(kind='hist')`.

4. **Box Plot (`kind='box'`)**: Box plots visualize the distribution of a dataset's values. You can create box plots using `df.plot(kind='box')`.

5. **Scatter Plot (`kind='scatter'`)**: Scatter plots are used to visualize the relationship between two variables. You can create scatter plots using `df.plot(kind='scatter')`.

6. **Pie Chart (`kind='pie'`)**: Pie charts represent data as a circular graph divided into slices, where each slice represents a proportion of the whole. You can create pie charts using `df.plot(kind='pie')`.

7. **Area Plot (`kind='area'`)**: Area plots are similar to line plots but filled under the line, creating an area. You can create area plots using `df.plot(kind='area')`.

8. **Hexbin Plot (`kind='hexbin'`)**: Hexbin plots are used for visualizing bivariate data with hexagonal bins. You can create hexbin plots using `df.plot(kind='hexbin')`.

9. **KDE Plot (`kind='kde'`)**: Kernel Density Estimation (KDE) plots estimate the probability density function of a continuous random variable. You can create KDE plots using `df.plot(kind='kde')`.

10. **Density Plot (`kind='density'`)**: Density plots are similar to KDE plots and represent the data's distribution. You can create density plots using `df.plot(kind='density')`.

11. **Bar Stacked Plot (`kind='bar'` with stacked parameter)**: Stacked bar plots show the composition of a category as stacked bars. You can create stacked bar plots using `df.plot(kind='bar', stacked=True)`.

12. **Line Stacked Plot (`kind='line'` with stacked parameter)**: Stacked line plots display multiple line plots stacked on top of each other. You can create stacked line plots using `df.plot(kind='line', stacked=True)`.

These are some of the common plot types available in Pandas. Depending on your data and visualization needs, you can choose the appropriate plot type to effectively convey your information.
'''


import pandas as pd
import matplotlib.pyplot as plt
filename = 'C:/Code Stuff/python-training/HW-07-10/Plot/NIFTY 50-08-10-2022-to-07-10-2023.csv'

df = pd.read_csv(filename, sep=',')
#Remove all whitespaces and info in the brackets like units etc.
df['Date'] = df['Date'].str.replace(r'(\d. )|[$@#]|\(.*\)',' ', regex=True)
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
df.plot(x='Date', y='Close', kind='line', title='Line Plot')
```

```python
df.plot(x='Date', y='Close', kind='bar', title='Bar Plot')
```

```python
#WITH MATPLOTLIB

# Define the bin size for the histogram
bin_size = 10

# Calculate the number of bins based on the bin size
num_bins = int((df['Close'].max() - df['Close'].min()) / bin_size)

# Create the histogram with specified bins
plt.hist(df['Close'], bins=num_bins, edgecolor='k', range=(df['Close'].min(), df['Close'].max()))

# Set labels and title
plt.xlabel('Close')
plt.ylabel('Frequency')
plt.title('Histogram with 100-Point Bins for Float Values')

# Show the histogram
plt.show()
```
